```python
import numpy as np
import matplotlib.pyplot as plt

# Create a sample data for t-theta waves
time = np.linspace(0, 10, 1000)
cell1 = np.sin(4 * np.pi * time) + np.sin(8 * np.pi * time)
cell2 = np.sin(6 * np.pi * time) + np.sin(12 * np.pi * time)

# Plot the t-theta waves for two cells
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].plot(time, cell1)
axs[0].set_title('Cell 1')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Amplitude')

axs[1].plot(time, cell2)
axs[1].set_title('Cell 2')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
