```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic neural data
sampling_rate = 2000
t = np.arange(0, 10, 1/sampling_rate)
theta = np.sin(2*np.pi*8*t) + np.sin(2*np.pi*8.2*t) + np.sin(2*np.pi*8.4*t)
noise = np.random.normal(0, 0.1, len(t))
signal = theta + noise

# Set threshold for spike detection
threshold = 0.5

# Find spike times
spike_times = []
for i in range(1, len(signal)-1):
    if signal[i] > threshold and signal[i] > signal[i-1] and signal[i] > signal[i+1]:
        spike_times.append(t[i])

# Plot signal and spikes
fig, ax = plt.subplots()
ax.plot(t, signal)
ax.plot(spike_times, np.ones(len(spike_times))*threshold, '|', markersize=15)
ax.set_xlabel('Time (s)')
ax.set_ylabel('Signal amplitude')
ax.set_title('Neural signal with detected spikes')
plt.show()
```

```python
import numpy as np

# Generate time vector
t = np.arange(0, 10, 0.001)

# Generate theta wave signal with added noise
theta = 5 * np.sin(2 * np.pi * 8 * t) + np.random.normal(0, 1, len(t))

# Set threshold for spike detection
threshold = 3

# Find times when signal crosses threshold and is higher than its two neighboring samples
spike_times = np.where((theta[1:-1] > threshold) & (theta[1:-1] > theta[:-2]) & (theta[1:-1] > theta[2:]))[0] + 1

# Print spike times
print("Spike times:", spike_times)

```

```python
from sklearn.cluster import KMeans

# Cluster theta wave signal into two clusters
kmeans = KMeans(n_clusters=2)
clusters = kmeans.fit_predict(theta.reshape(-1, 1))

# Visualize clustering results
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(t, theta, alpha=0.5)
ax.scatter(t, theta, c=clusters, cmap='viridis')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Theta wave signal')
ax.set_title('K-means clustering of theta wave signal')
plt.show()
```

```python
# Plot theta wave signal and detected spikes
fig, ax = plt.subplots()
ax.plot(t, theta, alpha=0.5)
ax.vlines(t[spike_times], ymin=0, ymax=np.max(theta), color='r', linewidth=2)
ax.set_xlabel('Time (s)')
ax.set_ylabel('Theta wave signal')
ax.set_title('Theta wave signal with detected spikes')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the theta wave generation
duration = 10  # Duration of the theta waves in seconds
sampling_rate = 1000  # Sampling rate in Hz
theta_freq = 8  # Frequency of the theta waves in Hz
amplitude = 1  # Amplitude of the theta waves

# Generate time axis
time = np.linspace(0, duration, int(duration * sampling_rate), endpoint=False)

# Generate theta waves
theta_waves = amplitude * np.sin(2 * np.pi * theta_freq * time)

# Plot the theta waves
plt.figure(figsize=(10, 4))
plt.plot(time, theta_waves)
plt.title('Synthetic Theta Waves')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
