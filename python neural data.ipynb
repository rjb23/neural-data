```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic neural data
sampling_rate = 2000
t = np.arange(0, 10, 1/sampling_rate)
theta = np.sin(2*np.pi*8*t) + np.sin(2*np.pi*8.2*t) + np.sin(2*np.pi*8.4*t)
noise = np.random.normal(0, 0.1, len(t))
signal = theta + noise

# Set threshold for spike detection
threshold = 0.5

# Find spike times
spike_times = []
for i in range(1, len(signal)-1):
    if signal[i] > threshold and signal[i] > signal[i-1] and signal[i] > signal[i+1]:
        spike_times.append(t[i])

# Plot signal and spikes
fig, ax = plt.subplots()
ax.plot(t, signal)
ax.plot(spike_times, np.ones(len(spike_times))*threshold, '|', markersize=15)
ax.set_xlabel('Time (s)')
ax.set_ylabel('Signal amplitude')
ax.set_title('Neural signal with detected spikes')
plt.show()
```
